```python
import pickle
import pyemma.coordinates as coor
import pyemma.msm as msm
import pyemma.plots as mplt
import numpy as np
from glob import glob
from copy import deepcopy
from msmbuilder.cluster import NDGrid
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
trajs = glob('./*.npy')
X = [np.load(x) for x in trajs]
```

```python
cluster = NDGrid(max=1.2, min=-1.2, n_bins_per_feature=60)
ctrajs_60 = cluster.fit_transform(X)
```

```python
cluster = NDGrid(max=1.2, min=-1.2, n_bins_per_feature=90)
ctrajs_90 = cluster.fit_transform(X)
```

```python
cluster = NDGrid(max=1.2, min=-1.2, n_bins_per_feature=220)
ctrajs_220 = cluster.fit_transform(X)
```

```python
mm = msm.BayesianMSM(lag=25, nsamples=500)
```

```python
models = []
for ctraj in [ctrajs_60, ctrajs_90, ctrajs_220]:
    mm = msm.BayesianMSM(lag=25, nsamples=500)
    models.append(mm.fit(ctraj))
```

```python
clusters = [60, 90, 220]
processes = [0,1,2]
k = len(processes)

nsamples = 500
```

```python
data = {'clusters': [], 'process': [], 'timescale': []}
for i, cluster in enumerate(clusters):
    allts = models[i].sample_f('timescales', k=3)
    allts = np.vstack(allts)
    data['clusters'].extend([cluster]*nsamples*k)
    for proc in processes:
        data['process'].extend([proc+2]*nsamples)
        data['timescale'].extend(allts[:, proc])
        
        
```

```python
for k, v in data.items():
    print(k, len(v))
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
exact_timescales = [844.4, 125.5, 64.3]
```

```python
with sns.plotting_context('paper', font_scale=1.5):
    sns.set_style('whitegrid')
    g = sns.FacetGrid(hue='clusters', row='process', data=df, sharex=False, sharey=False)
    g.map(sns.distplot, 'timescale', hist=False).add_legend()
    for i, ax in  enumerate(g.axes.flatten()):
        ax.vlines(exact_timescales[i], *ax.get_ylim(), )
        
    plt.savefig('../figures/cluster_vs_timescale_dist.png', transparent=True)
    
```
